```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
color_list = ['#60FAF8', '#97E3FE', '#92C2DD', '#F38E59', '#CC54D0', '#8B6A84', '#F394F8']
country_list = ['Ukraine', 'Russia', 'India', 'Cambodia', 'China', 'Poland', 'Germany']
marker_list = ['o', 's', 'v', '^', '+', '*', '+', '+']

df = pd.read_csv('2022-2023 TOP 8 NATIONS.csv',sep=',', header=None)
df.columns = ['Term', 'Country', 'Value']

fig, ax = plt.subplots(figsize = (16, 9))

for i, country in enumerate(country_list):
   df_country = df[df['Country'] == country]
   ax.plot(df_country['Term'], df_country['Value'], label=country, color = color_list[i]
            , marker=marker_list[i], linewidth = 2.5)

# ax.set_title('Frequency change chart of national words in the fifth quarter', fontsize = 26)
ax.set_xlabel('Terms', fontsize = 20)
ax.set_ylabel('Counts', fontsize = 20)


# 
plt.xticks(size=15)
plt.yticks(size=15)
plt.tight_layout()
# 
plt.legend(fontsize=20)
plt.show()

```

```python
color_list = ['#60FAF8', '#97E3FE', '#F8FE29', '#92C2DD', '#F38E59', '#CC54D0', '#8B6A84', '#F394F8']
tech_list = ["Phishing", "Email(attack)", "Ransomware", "DDos", "Backdoor"]
marker_list = ['o', 's', 'v', '^', '+']

tech = pd.read_csv('2022-2023 TECH KEYWORDS.csv',sep=',', header=None)
tech.columns = ['Term', 'Keywords', 'Value']

fig, ax = plt.subplots(figsize = (16, 9))

for i, tech_i in enumerate(tech_list):
   df_tech = tech[tech['Keywords'] == tech_i]
   ax.plot(df_tech['Term'], df_tech['Value'], label=tech_i, color = color_list[i]
            , marker=marker_list[i], linewidth = 2.5)

# ax.set_title('Frequency change chart of national words in the fifth quarter', fontsize = 26)
ax.set_xlabel('Terms', fontsize = 20)
ax.set_ylabel('Counts', fontsize = 20)


# 
plt.xticks(size=15)
plt.yticks(size=15)
plt.tight_layout()
# 
plt.legend(fontsize=20)
plt.show()
```
